## Survial Rate of Passanger in titanic

```python
import pandas as pd
df = pd.read_csv('titanic.csv')
df.head()
```

#### Droping the columns that we don't need 

```python
df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'], axis = 'columns', inplace = True)
```

```python
df.head()
```

#### Targeting the survial rate of Pasanger

```python
inputs = df.drop('Survived', axis= 'columns')
target = df.Survived
```

```python
inputs.head()
```

```python
inputs.Age = inputs.Age.fillna(inputs.Age.mean())
```

Labling the sex column

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_sex = LabelEncoder()
```

```python
inputs['Sex_n'] = le_sex.fit_transform(inputs['Sex'])
inputs_n = inputs.drop(['Sex'], axis = 'columns')
```

```python
inputs_n.head()
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
model.fit(inputs_n, target)
```

Let's predict the Death Rate of passanger whose id is pclass is 1, age 35, Fare 53.100, and sex is female.

```python
model.predict([['1','35','53.1','0']])
```

score of model

```python
model.score(inputs_n,target)
```
